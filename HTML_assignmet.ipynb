##Question 1  :-  Write a simple program in HTML  that displays the heading "HTML defines the content and structure of your website” on the web browser?

Answer :- Here is a simple HTML program that displays the heading "HTML defines the content and structure of your website" on a web browser:  

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML Basics</title>
</head>
<body>
    <h1>HTML defines the content and structure of your website</h1>
</body>
</html>
```

### Explanation:
- The `<!DOCTYPE html>` declaration defines the document as an HTML5 document.
- The `<html>` element is the root element of the page.
- The `<head>` section contains metadata and the title of the webpage.
- The `<body>` section contains the visible content, where the `<h1>` tag is used to display the heading.

##Question 2 :-  Explain the purpose of comments in HTML and provide an example of how to use comments in an HTML document ?

Answer :- Purpose of Comments in HTML :

Comments in HTML are used to add notes or explanations within the code that are not displayed on the web page. They help improve code readability and maintainability by providing descriptions, reminders, or temporary disabling of code.

### **Uses of Comments in HTML**
1. **Documentation** – Explain the purpose of code for future reference.
2. **Debugging** – Temporarily disable a section of code.
3. **Collaboration** – Help other developers understand the code structure.

### **Syntax of HTML Comments**
```html
<!-- This is a comment -->
```

### **Example of Using Comments in an HTML Document**
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML Comments Example</title>
</head>
<body>
    <!-- This is a heading for the web page -->
    <h1>Welcome to My Website</h1>

    <!-- Below is a paragraph explaining the website -->
    <p>This website provides tutorials on web development.</p>

    <!-- Temporarily disabling an image -->
    <!-- <img src="image.jpg" alt="Sample Image"> -->
</body>
</html>
```

### **Key Points**
- HTML comments start with `<!--` and end with `-->`.
- They do **not** appear on the webpage but are visible in the page source code.
- They are useful for documentation and debugging without affecting the layout.

##Question 3 :-  Write an HTML program that includes a heading, a paragraph of text, a horizontal line, and a line break.Arrange these elements to create a simple web page layout ?

Answer :- Here is a simple HTML program that includes a heading, a paragraph, a horizontal line, and a line break to create a basic web page layout:  

```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Simple Web Page</title>
</head>
<body>

    <!-- Heading -->
    <h1>Welcome to My Web Page</h1>

    <!-- Paragraph -->
    <p>This is a simple web page created using HTML. It contains a heading, a paragraph, a horizontal line, and a line break.</p>

    <!-- Horizontal Line -->
    <hr>

    <!-- Another paragraph with a line break -->
    <p>HTML is the standard markup language for creating web pages.<br>
    It defines the structure and content of a webpage.</p>

</body>
</html>
```

### **Explanation of Elements Used**
1. `<h1>` – Displays the main heading.
2. `<p>` – Defines a paragraph of text.
3. `<hr>` – Inserts a horizontal line to separate sections.
4. `<br>` – Inserts a line break within the paragraph.

### **How It Looks in a Browser**
- The heading appears at the top.
- A paragraph follows below it.
- A horizontal line separates the content.
- A second paragraph with a line break is displayed after the horizontal line.

##Question 4 :- Write a short note on Tag and element with an example?

Answer :- Tag vs. Element in HTML :  

#### **1. HTML Tag**  
A **tag** is a keyword enclosed in angle brackets (`<>`) that defines the beginning or end of an HTML element. Tags are used to create elements and structure content on a webpage.  

**Example of Tags:**  
- `<p>` (opening tag)  
- `</p>` (closing tag)  
- `<br>` (self-closing tag)  

#### **2. HTML Element**  
An **element** consists of an opening tag, content (if applicable), and a closing tag. It represents a part of the webpage, such as headings, paragraphs, or images.  

**Example of an Element:**  
```html
<p>This is a paragraph.</p>
```
- `<p>` is the opening tag.
- `"This is a paragraph."` is the content.
- `</p>` is the closing tag.
- Together, they form a **paragraph element**.  

### **Self-Closing Elements**  
Some elements do not have closing tags; they close themselves.  
Example:  
```html
<br>  <!-- Line break element -->
<hr>  <!-- Horizontal line element -->
<img src="image.jpg" alt="Sample Image">  <!-- Image element -->
```

#### **Summary**  
- **Tags** are used to define elements.
- **Elements** include the tag and its content (if any).  
- **Some elements** are self-closing and do not require a closing tag.

##Question 5 :-  What is the DOCTYPE Declaration in HTML?

Answer :- DOCTYPE Declaration in HTML :

The **DOCTYPE declaration** is an instruction that tells the web browser which version of HTML is being used in the document. It ensures that the browser renders the webpage correctly.  

### **Syntax of DOCTYPE in HTML5**  
```html
<!DOCTYPE html>
```
- This declaration must be placed at the **very beginning** of an HTML document.
- It is **not case-sensitive**, but it is usually written in uppercase.
- It does **not** have a closing tag.

### **Purpose of DOCTYPE**
1. **Defines the Document Type** – Informs the browser that the document is written in **HTML5**.
2. **Ensures Standard Rendering** – Helps browsers display the webpage in **standards mode** instead of **quirks mode** (which can cause inconsistent rendering).
3. **Improves Compatibility** – Helps modern and older browsers interpret the webpage properly.

### **Example of an HTML5 Document with DOCTYPE**
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>HTML5 DOCTYPE Example</title>
</head>
<body>
    <h1>Hello, World!</h1>
    <p>This webpage follows the HTML5 standard.</p>
</body>
</html>
```

### **Older DOCTYPE Versions (For Reference)**
- **HTML 4.01 (Strict)**
  ```html
  <!DOCTYPE HTML PUBLIC "-//W3C//DTD HTML 4.01//EN" "http://www.w3.org/TR/html4/strict.dtd">
  ```
- **XHTML 1.0 (Strict)**
  ```html
  <!DOCTYPE html PUBLIC "-//W3C//DTD XHTML 1.0 Strict//EN" "http://www.w3.org/TR/xhtml1/DTD/xhtml1-strict.dtd">
  ```

#### **Conclusion**
- The DOCTYPE declaration is essential for proper rendering.
- HTML5 uses the simple `<!DOCTYPE html>`, making it easier to use than older versions.
- Always include DOCTYPE at the beginning of an HTML document.
